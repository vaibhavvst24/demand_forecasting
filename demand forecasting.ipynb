```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Load your data from the CSV file (replace with your actual file path)
data = pd.read_csv(r"C:\Users\vaibh\Downloads\demand data 2.csv")

# Extract the actual values (y_true) and predicted values (y_pred) from the DataFrame
y_true = data['Unit Sales'].values  
y_pred = data['consumer confidence'].values  

# Calculate Mean Absolute Error (MAE)
mae = mean_absolute_error(y_true, y_pred)
print(f'Mean Absolute Error (MAE): {mae}')

# Calculate Root Mean Squared Error (RMSE)
rmse = np.sqrt(mean_squared_error(y_true, y_pred))
print(f'Root Mean Squared Error (RMSE): {rmse}')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load retail sales data from a CSV file (replace with your file path)
data = pd.read_csv(r"C:\Users\vaibh\Downloads\demand data 2.csv")

# Convert the date column to a datetime object
data['instant'] = pd.to_datetime(data['instant'])

# Set the date column as the index (optional but can be helpful for time series analysis)
data.set_index('instant', inplace=True)

# Plot retail sales over time to visualize seasonal trends
plt.figure(figsize=(10, 4))
plt.plot(data.index, data['Unit Sales'], label='Retail Sales', color='blue')
plt.xlabel('instant')
plt.ylabel('Unit Sales')
plt.title('Retail Sales Over Time')
plt.grid(True)
plt.legend()

# Create subplots to visualize economic indicators
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(12, 10))
data.plot(y='GDP', ax=axes[0], label='GDP', color='green')
axes[0].set_xlabel('instant')
axes[0].set_ylabel('GDP')
axes[0].set_title('GDP Over Time')

data.plot(y='inflation_ rate', ax=axes[1], label='Inflation Rate', color='red')
axes[1].set_xlabel('instant')
axes[1].set_ylabel('inflation_ rate')
axes[1].set_title('Inflation Rate Over Time')

data.plot(y='consumer confidence', ax=axes[2], label='Consumer Confidence', color='purple')
axes[2].set_xlabel('Date')
axes[2].set_ylabel('consumer confidence')
axes[2].set_title('Consumer Confidence Over Time')

plt.tight_layout()

# Show the plots
plt.show()
```
